```python
from tabulate import tabulate


def print_table(dataset):
    table = [[''] * 3 for i in range(3)]
    for i in range(3):
        for j in range(3):
            table[i][j] = (dataset[(3*i) + j]) 
    
    print(tabulate(table, tablefmt='grid'))

    
def wins(dataset):
    win_set = {(0,3,6), (1,4,7), (2,5,8), (0,1,2), (3,4,5), (6,7,8), (0,4,8), (2,4,6)}
    empty_spaces = [i for i in range(9) if dataset[i] == '']
    
    for i in win_set:
        if (dataset[i[0]] == dataset[i[1]] == dataset[i[2]]) and (dataset[i[0]] == 'X' or dataset[i[0]] == 'O'):
            return dataset[i[0]]
    
    if len(empty_spaces) == 0:
        return "D"
    else:
        return 0

  
    
def player_user(board):
    print_table(board)
    try:
        turn = int(input("Enter the cell number of the patient, you want to apply your treatment : "))
    except ValueError:
        print("\nPlease enter an integer value from 1 to 9")
        return player_user(board)
    
    if (turn < 1) or (turn > 9):
        print("\nEnter an integer value from 1 to 9")
        return player_user(board)
    elif (board[turn - 1] != ''):
        print("\nThe number you entered has already been used!")
        return player_user(board)
    else:
        board[turn - 1] = 'O'
        return board
    
    

def minimax(board, depth, is_maximizing):
    status = wins(board)
    if status == 'X':
        return 100
    elif status == 'O':
        return -100
    elif status == 'D':
        return 0

    if is_maximizing:
        best_score = -1000
        for key in range(9):
            if board[key] == '':
                board[key] = 'X'
                score = minimax(board, depth + 1, False)
                board[key] = ''
                if score > best_score:
                    best_score = score               
                
        return best_score

    else:
        best_score = 1000
        for key in range(9):
            if board[key] == '':
                board[key] = 'O'
                score = minimax(board, depth + 1, True)
                board[key] = ''
                if score < best_score:
                    best_score = score 
        return best_score


def player_AI(board):
    best_score = -1000
    best_move = 0

    for key in range(9):
        if board[key] == '':
            board[key] = 'X'
            score = minimax(board, 0, False)
            board[key] = ''
            if score > best_score:
                best_score = score
                best_move = key
    board[best_move] = 'X'
    return board

    

def play_game(first_turn):
    board = ['','','','','','','','','']
    turns = 9
    AI = True
    if first_turn == True:
        AI = False
        
    
    for i in range(turns):
        if AI == True:
            
            print("\n -------------------AI's turn--------------------")
            board = player_AI(board)
            print_table(board)
            check = wins(board)
            if check == 'X':
                print(check)
                return "UNFORTUNATELY! the treatment couldn't stop the spread of cancer"
            AI = False
        else:
            
            print("\n -------------------User's turn--------------------")
            board = player_user(board)
            print_table(board)
            check = wins(board)
            if check == 'O':
                print(check)
                return "CONGRATULATIONS! you (player O) successfully treated cancer"
            AI = True
      
    return "You managed to provide right treatment for the cancer cells but were not able to treat it"  
        

print("Welcome to the AI-Driven Cancer Treatment Decision Simulator")
print("============================================================\n\n")
print("The game board below represents the patient profile – a 3x3 grid where each cell corresponds to a specific aspect of the patient's condition\n")
print_table(['','','','','','','','',''])
print("\nThe initial state of the board reflects the patient's current health status.")
print("\n\n There are numbers assigned to each cell as following:")
print_table(['1','2','3','4','5','6','7','8','9'])
print("\n\nThe simulation will work by the doctor and AI making moves in alternative turns")
print("The Doctor (user) takes turns choosing a treatment option which will the available options from 1 to 9")
print("Meanwhile the Cancer (AI opponent) reacts to each treatment decision made by the doctor(user)")
print("\n\nYou can use the below grid as a reference:\n\n")
print_table(['1','2','3','4','5','6','7','8','9'])
print("\n\n")

player_first = input("Would you like to to take the first turn, Enter 'Y' for Yes and 'N' for No:  ")

print("=====================  Game Begins  =============================")
print("\n You are 'O' and AI is 'X'\n\n")
if player_first == 'y' or player_first == 'Y':
    result = play_game(True)
else:
    result = play_game(False)

print("\n\n",result)
```
